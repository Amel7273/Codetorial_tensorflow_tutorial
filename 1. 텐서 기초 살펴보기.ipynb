### 3) 텐서 만들기

```python
# 모듈 임포트
import tensorflow as tf
```

```python
# tf.constant() 사용하기
a = tf.constant(1)
b = tf.constant([2])
c = tf.constant([[1,2],[3,4]])
d = tf.constant([[[1,2],[3,4]],[[5,6],[7,8]]])

print(a)
print(b)
print(c)
print(d)
```

tf.constant()는 상수 텐서를 만듭니다.

```python
# tf.zeros()사용하기

a = tf.zeros(1)
b = tf.zeros([2])
c = tf.zeros([2,3])

print(a)
print(b)
print(c)
```

tf.zeors()는 요소가 0인 텐서를 만듭니다.

```python
#tf.ones() 사용하기

a = tf.ones(3)
b = tf.ones([4])
c = tf.ones([2,2,3])

print(a)
print(b)
print(c)
```

tf.ones ()는 요소가 1인 텐서를 만듭니다.

```python
#tf.range() 사용하기

a = tf.range(0,3)
b = tf.range(1,5,2)

print(a)
print(b)
```

tf.range()는 파이썬 range()와 비슷하게, 주어진 범위와 간격을 갖는 숫자들의 시퀀스를 만듭니다.

```python
# tf.linspace() 사용하기

a = tf.linspace(0,1,3)
b = tf.linspace(0,3,10)

print(a)
print(b)
```

tf.linspace()는 numpy.linspace()와 비슷하게, 주어진 범위를 균일한 간격으로 나누는 숫자의 시퀸스를 반환합니다.

### 4) 텐서의 자료형과 형태

```python
# 예제 1

a = tf.constant(1)
b = tf.constant([2])
c = tf.constant([3,4,5])

print(a.dtype, a.shape)
print(b.dtype, b.shape)
print(c.dtype, c.shape)
```

dtype, shape 속성은 각각 텐서(tensor)의 자료형과 형태를 반환합니다.

텐서 a,b,c는 모두 정수형(int32)의 데이터를 갖고, 각각(),(1,),(3,)의 형태를 가짐을 알 수 있다.

```python
# 예제 2

d = tf.constant([1., 2.])
e = tf.constant([[1, 2. ,3]])
f = tf.constant([[1,2], [3,4]])

print(d.dtype, d.shape)
print(e.dtype, e.shape)
print(f.dtype, f.shape)
```

텐서 d는 실수형(float32)의 데이터를 갖고, (2,)의 형태를 가집니다.
텐서 e는 실수형(float32)의 데이터를 갖고, (1,3)의 형태를 가집니다.
텐서 f는 정수형(int32)의 데이터를 갖고, (2,2)의 형태를 가집니다.

### 5) 간단한 수학 연산 



```python
a = tf.add(1, 2)
b = tf.subtract(10, 5)
c = tf.square(3)
d = tf.reduce_sum([1, 2, 3])
e = tf.reduce_mean([1, 2, 3])

print(a)
print(b)
print(c)
print(d)
print(e)
```

TensorFlow는 텐서 (Tensor)를 생성하고 연산하는 다양한 수학적 함수를 제공합니다.

### 6) Numpy 호환성

텐서(Tensor)는 Numpy Array와 비슷하지만,

* 텐서는 GPU, TPU와 같은 가속기에서 사용할 수 있고
* 텐서는 값을 변경할 수 없습니다.

```python
# 텐서를 Numpy 어레이로
a = tf.constant([1,2,3])

print(a)
print(a.numpy())
```

텐서의 **numpy()** 메서드를 이용해 텐서를 Numpy Array로 변환할 수 있습니다.

```python
# Numpy 어레이를 텐서로

import numpy as np

b = np.ones(3)

print(b)
print(tf.multiply(b,2)) # 다른 tensorflow 메서드도 가능
```

다양한 Tensorflow연산을 Numpy Array에 대해 수행하기만 한다면 자동으로 넘파이를 텐서로 바꿀 수 있습니다.
